```python
import fastai
import torch
```

```python
from fastai.tabular.all import *

# dataset
path = untar_data(URLs.ADULT_SAMPLE)

# dataloaders

dls = TabularDataLoaders.from_csv(path/'adult.csv',path=path,bs=64,y_names='salary', skipinitialspace=False,cat_names=['workclass','education','marital-status','occupation','relationship','race','sex'],
                                cont_names=['age','fnlwgt','education-num','capital-gain','capital-loss','hours-per-week'],
                                procs = [Categorify,FillMissing,Normalize])


# train

learn = tabular_learner(dls,metrics=accuracy)
learn.fit_one_cycle(4)
```

```python
# loading data set into dataframe

import pandas as pd
df = pd.read_csv(path/'adult.csv')
df.head(10)
```

```python
# test
row, pred, probs = learn.predict(df.iloc[4])
if pred.item()==1:
  print("The salary >= 50K")
elif pred.item()==0:
  print('The salary < 50K')
pred.item(), probs
```
